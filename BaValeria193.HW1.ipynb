```python
import cmath
import numpy as np
def solve_quad(b,c):
    Dis = b*b - 4*c
    if Dis >= 0:
        x1 = (-b + np.sqrt(Dis)) / (2)
        x2 = (-b - np.sqrt(Dis)) / (2)        
    else:
        x1 = (-b + cmath.sqrt(Dis)) / (2)
        x2 = (-b - cmath.sqrt(Dis)) / (2)
    v = (x1, x2)
    return (v)



from numpy import allclose
variants = [{'b': 4.0, 'c': 3.0},
            {'b': 2.0, 'c': 1.0},
            {'b': 0.5, 'c': 4.0},
            {'b': 1e10, 'c': 3.0},
            {'b': -1e10, 'c': 4.0},]

for v in variants:
    x1, x2 = solve_quad(**v)
    print(allclose(x1*x2, v['c']))
```
